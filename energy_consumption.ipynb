```python
import matplotlib.pyplot as plt

# Function to input monthly energy consumption data
def input_energy_data():
    data = []
    months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
    for month in months:
        print(f"Enter energy consumption data for {month}:")
        electricity = float(input("Electricity usage (kWh): "))
        gas = float(input("Gas usage (cubic meters): "))
        water = float(input("Water usage (cubic meters): "))
        data.append({'Month': month, 'Electricity': electricity, 'Gas': gas, 'Water': water})
    return data
```

1. import matplotlib.pyplot as plt: this line perticulary imports python data that is used for plotting and graphing data.
2. def input_energy_data(): is used to define line and similar reusable code blocks, functions to carry out perticluar responsibilities.
3. data=[ ]: it is used to initialise calleddata by line.
4. months= ['january' , 'february' ..... 'december']: this line generates list of tewlve months.
5. print(f"Enter energy consumption data for {month}:"): This line prompts the user to enter their energy consumption.
overall with the help of this code, the user can enter monthly statistics on their usage of gas, electricity, and water for a full year. It keeps this information in a list of dictionaries, each of which shows the amount of energy used in a certain month.


```python
# Function to visualize monthly consumption trends
def visualize_monthly_trends(data):
    months = [entry['Month'] for entry in data]
    electricity = [entry['Electricity'] for entry in data]
    gas = [entry['Gas'] for entry in data]
    water = [entry['Water'] for entry in data]

    plt.figure(figsize=(10, 5))
    plt.plot(months, electricity, label='Electricity', marker='o')
    plt.plot(months, gas, label='Gas', marker='o')
    plt.plot(months, water, label='Water', marker='o')
    plt.title('Monthly Energy Consumption Trends')
    plt.xlabel('Month')
    plt.ylabel('Consumption')
    plt.legend()
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

1. def visualize_monthly_trends(data): this code visulise the monthly consumption trends of energy. moreover, it initiates to output in the way user wants with regard to figure size, label, layout and so on.

```python
# Function to compare energy types by consumption
def compare_energy_types(data):
    total_electricity = sum(entry['Electricity'] for entry in data)
    total_gas = sum(entry['Gas'] for entry in data)
    total_water = sum(entry['Water'] for entry in data)

    categories = ['Electricity', 'Gas', 'Water']
    consumption = [total_electricity, total_gas, total_water]
    plt.figure(figsize=(7, 5))
    plt.bar(categories, consumption, color=['blue', 'orange', 'green'])
    plt.title('Energy Consumption by Type')
    plt.xlabel('Energy Type')
    plt.ylabel('Total Consumption')
    plt.grid(axis='y')
    plt.tight_layout()
    plt.show()

# Function to show annual total consumption breakdown by energy type
def annual_consumption_breakdown(data):
    total_electricity = sum(entry['Electricity'] for entry in data)
    total_gas = sum(entry['Gas'] for entry in data)
    total_water = sum(entry['Water'] for entry in data)

    categories = ['Electricity', 'Gas', 'Water']
    consumption = [total_electricity, total_gas, total_water]

    plt.pie(consumption, labels=categories, autopct='%1.1f%%', startangle=140)
    plt.title('Annual Energy Consumption Breakdown')
    plt.axis('equal')
    plt.tight_layout()
    plt.show()
```

1. the function comparison is called to inorder to compare energy such as electricity, water and gas sonsumption.
2. The figure size has been enlarge to 7 in with and 5 inches form standard size that is 6.4 in with and 4.8(inches) in height.
3. plt.bar command is being used to along with different colours to differentiate between energy consumption and catagory.
4. The analysis and visualisation of annual energy consumption data is essentially captured by the annual_consumption_breakdown function. The information is presented in a pie chart format, which makes it simple to understand how energy usage is distributed among various categories.

```python

def main():
    print("Welcome to the Energy Consumption Analysis Tool")
    data = input_energy_data()

    while True:
        print("\nSelect an analysis option:")
        print("1. Monthly Consumption Trends")
        print("2. Comparison of Energy Types by Consumption")
        print("3. Annual Total Consumption Breakdown by Energy Type")
        print("4. Exit")
        choice = input("Enter your choice (1-4): ")

        if choice == '1':
            visualize_monthly_trends(data)
        elif choice == '2':
            compare_energy_types(data)
        elif choice == '3':
            annual_consumption_breakdown(data)
        elif choice == '4':
            print("Thank you for using the Energy Consumption Analysis Tool. Goodbye!")
            break
        else:
            print("Invalid choice. Please enter a number between 1 and 4.")

if __name__ == "__main__":
    main()
```
